```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Setup Cavity Parameters
num_modes = 50          # Number of longitudinal modes
t = np.linspace(0, 10, 4000)  # Time axis
frequencies = np.arange(10, 10 + num_modes) # Equally spaced mode frequencies

# 2. CW Simulation (Random Phases)
# Every mode has a random phase shift between 0 and 2*pi
random_phases = np.random.uniform(0, 2 * np.pi, num_modes)
cw_signal = np.zeros_like(t)
for i, f in enumerate(frequencies):
    cw_signal += np.cos(2 * np.pi * f * t + random_phases[i])

# 3. Mode-Locked Simulation (Fixed Phases)
# Every mode has the EXACT same phase (locked at 0)
locked_phases = np.ones(num_modes) * 86.23874  # Arbitrary fixed phase
ml_signal = np.zeros_like(t)
for i, f in enumerate(frequencies):
    ml_signal += np.cos(2 * np.pi * f * t + locked_phases[i])

# 4. Visualization
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, cw_signal**2, color='blue', alpha=0.7)
plt.title(f"CW Laser Output (Random Phases among {num_modes} modes)")
plt.ylabel("Intensity (E)")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(t, ml_signal**2, color='red')
plt.title(f"Mode-Locked Laser Output (Locked Phases among {num_modes} modes)")
plt.xlabel("Time")
plt.ylabel("Intensity (E)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
